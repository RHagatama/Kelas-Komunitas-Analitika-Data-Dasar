```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
```

# Visualisasi Relasi

```python
# Load dataset
mpg = sns.load_dataset('mpg')
mpg
```

```python
# Simple scatterplot
sns.relplot(data=mpg, x='weight', y='mpg');
```

```python
# Add third variable (hue)
sns.relplot(data=mpg, x='weight', y='mpg', hue='origin');
```

```python
# Add third variable (style)
sns.relplot(data=mpg, x='weight', y='mpg', hue='acceleration', style='origin');
```

```python
# Add third variable (style)
sns.relplot(data=mpg, x='weight', y='mpg', size='origin');
```

```python
# Visualisasi data runtun waktu
sns.relplot(data=mpg, x='model_year', y='mpg', kind='line', ci=None);
```

```python
# Visualisasi data runtun waktu and add third variable
sns.relplot(data=mpg, x='model_year', y='mpg', hue='origin', kind='line', ci=None);
```

# Visualisasi Distribusi

```python
# Load dataset
diamonds = sns.load_dataset('diamonds')
diamonds
```

```python
# Distribusi univariat
sns.displot(data=diamonds, x='carat');
```

```python
# Mengatur rentang bins
sns.displot(data=diamonds, x='carat', binwidth=0.01);
```

```python
# Mengatur jumlah bins
sns.displot(data=diamonds, x='carat', bins=30);
```

```python
# Membandingkan distribusi
sns.displot(data=mpg, x='mpg', hue='origin');
```
