## 2. Filter

```python
# Memuat Data
flights = pd.read_csv('https://raw.githubusercontent.com/RHagatama/Kelas-Komunitas-Analitika-Data-Dasar/main/flights.csv')
flights
```

### 2.1. Memilih kolom

```python
# Memilih kolom
kolom = ['year', 'month', 'day']
flights[kolom]
```

### Exercise

1. Pilihlah kolom ini dari dataframe: 'month', 'day', 'dep_time', 'dep_delay', 'arr_time', 'arr_delay', 'carrier', 'origin', 'dest', 'air_time', 'distance'
2. Simpan hasilnya ke dalam objek bernama flights

```python
kolom = ['month', 'day', 'dep_time', 'dep_delay', 'arr_time', 'arr_delay', 'carrier', 'origin', 'dest', 'air_time', 'distance']
flights = flights[kolom]
flights
```

### 2.2. Filter Baris 

```python
# Memilih penerbangan di bulan Juni
juni = flights['month'] == 6
flights[juni]
```

```python
# Memilih penerbangan dengan jarak penerbangan di atas 1000 mil
far_distance = flights['distance'] > 1000
flights[far_distance]
```

```python
# Memilih penerbangan di 1 minggu pertama per bulannya
first_week = flights['day'] <= 7
flights[first_week]
```

```python
# Memilih penerbangan pada 1 Januari
jan_1 = (flights['month'] == 1) & (flights['day'] == 1)
flights[jan_1]
```

```python
# Memilih penerbangan dari Juli hingga September
jul_to_sep = (flights['month'] >= 7) & (flights['month'] <= 9)
flights[jul_to_sep]
```

```python
# Memilih penerbangan dari bulan November hingga Januari
nov_to_jan = (flights['month'] >= 11) | (flights['month'] == 1)
flights[nov_to_jan]
```

```python
# Memilih penerbangan yang tidak telat
not_late = ~(flights['arr_delay'] > 0)
flights[not_late]
```

```python
# Memilih penerbangan dengan maskapai DL atau UA
DL_UA = (flights['carrier'] == 'DL') | (flights['carrier'] == 'UA')
flights[DL_UA]
```

```python
# Memilih penerbangan dengan maskapai DL, AA, B6 atau UA
maskapai = ['DL', 'AA', 'B6', 'UA']
kondisi = flights['carrier'].isin(maskapai)
flights[kondisi]
```

### 2.3. Mengurutkan Baris

```python
# Mengurutkan dari bulan terkecil ke terbesar
flights.sort_values('month')
```

```python
# Mengurutkan berdasarkan tanggal dan bulan
flights.sort_values(['month', 'day'])
```

```python
# Mengurutkan penerbangan dari delay terbesar
flights.sort_values('arr_delay', ascending=False)
```

```python
# Mengurutkan dari jarak terdekat dan delay terlama
flights.sort_values(['distance', 'arr_delay'], ascending=[True, False])
```

```python
# Mengambil 3 penerbangan paling telat
flights.nlargest(3, 'arr_delay')
```

```python
# Mengambil 3 penerbangan paling dini
flights.nsmallest(3, 'arr_delay')
```

### 2.4. Bikin Kolom Baru

```python
# Menambahkan kolom arr_delay dalam satuan jam
flights.assign(
    arr_delay_hour = flights['arr_delay'] / 60,
    dep_delay_hour = flights['dep_delay'] / 60
)
```

```python
# Menggunakan if else
flights.assign(
    telat = np.where(flights['arr_delay'] > 0, True, False),
    jarak = np.where(flights['distance'] > 1000, 'jauh',
            np.where(flights['distance'] > 500, 'standar',
            np.where(flights['distance'] <= 500, 'dekat', 
            'tidak tahu')))
)
```

### 2.5. Agregasi Data

```python
# Mencari rata-rata jarak penerbangan
flights['distance'].mean()
```

```python
flights.describe()
```

```python
# Berdasarkan bulan hitung rata-rata keterlambatan
flights.groupby('month')['arr_delay'].mean()
```

```python
# Berdasarkan maskapai hitung rata-rata keterlambatan
flights.groupby('carrier').agg({
    'carrier':'count',
    'arr_delay':'mean',
    'distance':'mean'
})
```
