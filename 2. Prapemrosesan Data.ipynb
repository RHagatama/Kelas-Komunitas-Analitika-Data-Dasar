## 1. Memuat Data

```python
import pandas as pd
import numpy as np
```

```python
flights_raw = pd.read_csv('flights.csv', index_col=0)
flights_raw.head()
```

## 2. Memilih Kolom

```python
kolom = ['year', 'month', 'day']
flights_raw[kolom]
```

```python
kolom = ['month', 'day', 'dep_time', 'dep_delay', 'arr_time', 'arr_delay', 'carrier', 'origin', 'dest', 'air_time', 'distance']
flights = flights_raw[kolom]
flights.head()
```

## 3. Filter Baris

```python
# Memilih penerbangan di bulan Juni
juni = flights['month'] == 6
flights[juni]
```

```python
# Memilih penerbangan dengan jarak penerbangan di atas 1000 km
thousand_km = flights['distance'] >= 1000
flights[thousand_km]
```

```python
# Memilih penerbangan di tiap minggu pertama
first_week = flights['day'] <= 7
flights[first_week]
```

```python
# Memilih penerbangan pada 1 januari
first_jan = (flights['month'] == 1) & (flights['day'] == 1)
flights[first_jan]
```

```python
# Memilih penerbangan dari bulan Juli hingga September
jul_to_sep = (flights['month'] >= 7) & (flights['month'] <= 9)
flights[jul_to_sep]
```

```python
# Memilih penerbangan pada bulan November hingga Januari
nov_to_jan = (flights['month'] >= 11) | (flights['month'] == 1)
flights[nov_to_jan]
```

```python
# Memilih penerbangan berangkat yang tidak telat
not_delayed = ~(flights['dep_delay'] > 0)
flights[not_delayed]
```

```python
# Memilih penerbangan dengan maskapai DL dan UA
maskapai = flights['carrier'].isin(['DL', 'UA'])
flights[maskapai]
```

## 4. Missing Value

```python
# Menemukan waktu keberangkatan dengan Missing Value
na = flights['dep_time'].isna()
flights[na]
```

```python
# Memfilter missing value
flights.dropna()
```

```python
# Menyingkirkan missing value hanya dari kolom tertentu
not_na = flights['dep_delay'].notna()
flights[not_na]
```

## 6. Membuat kolom baru

```python
# Membuat kolom dep_delay dalam satuan detik dan jarak dalam satuan meter
flights = flights.assign(dep_delay_second = flights['dep_delay'] * 60,
                         distance_in_meter = flights['distance'] * 1000)

flights.head()
```

```python
# Menggunakan if else
flights = flights.assign(semester = np.where(flights['month'] <= 6, 1,
                                    np.where(flights['month'] >= 7, 2, 0)))

flights.head()
```

```python
# Menghapus kolom
flights.drop(columns=['dep_delay_second', 'semester', 'distance_in_meter'])
```

## 7. Agregasi Data

```python
# Menghitung rata-rata keterlambatan penerbangan
flights['dep_delay'].mean()
```

```python
# Menghitung rata-rata keterlambatan penerbangan berdasarkan bulan
flights.groupby('month')['dep_delay'].mean()
```

```python
# Menghitung jumlah penerbangan, rata-rata jarak penerbangan, dan rata-rata durasi keterlambatan berdasarkan tujuan
flights.groupby(['dest']).agg({
    'dest':'count',
    'distance':'mean',
    'arr_delay':'mean'
})
```
