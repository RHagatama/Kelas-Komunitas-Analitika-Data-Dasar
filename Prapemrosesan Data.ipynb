# Prapemrosesan Data

## 1. Memuat Data

```python
import pandas as pd
import numpy as np
```

```python
flights = pd.read_csv('flights.csv', index_col=0)
flights.head()
```

## 2. Memilih Kolom

```python
kolom = ['year', 'month', 'day']
flights[kolom]
```

## 3. Filter Baris

```python
# Memilih penerbangan di bulan Juni
juni = flights['month'] == 6
flights[juni]
```

```python
# Memilih penerbangan dengan jarak penerbangan di atas 1000 km
thousand_km = flights['distance'] >= 1000
flights[thousand_km]
```

```python
# Memilih penerbangan di tiap minggu pertama
first_week = flights['day'] <= 7
flights[first_week]
```

```python
# Memilih penerbangan pada 1 januari
first_jan = (flights['month'] == 1) & (flights['day'])
flights[first_jan]
```

```python
# Memilih penerbangan dari bulan Juli hingga September
jul_to_sep = (flights['month'] >= 7) & (flights['month'] <= 9)
flights[jul_to_sep]
```

```python
# Memilih penerbangan dengan tujuan 'IAH' atau 'HOU'
houston = (flights['dest'] == 'IAH') | (flights['dest'] == 'HOU')
flights[houston]
```

```python
# Memilih penerbangan berangkat yang tidak telat
not_delayed = ~(flights['dep_delay'] > 0)
flights[not_delayed]
```

```python
# Memilih penerbangan dengan maskapai DL dan UA
maskapai = flights['carrier'].isin(['DL', 'UA'])
flights[maskapai]
```

## 4. Missing Value

```python
# Menemukan waktu keberangkatan dengan Missing Value
na = flights['dep_time'].isna()
flights[na]
```

```python
# Memfilter missing value
flights[~na]
```

## 5. Mengurutkan Nilai

```python
# Mengurutkan penerbangan dari yang paling dini
flights.sort_values('dep_time')
```

```python
# Mengurutkan penerbangan dari yang paling akhir
flights.sort_values('dep_time', ascending=False)
```

```python
# Mengambil 3 penerbangan paling terlambat
flights.nlargest(3, 'dep_delay')
```

```python
# Mengambil 3 penerbangan paling dini
flights.nsmallest(3, 'dep_delay')
```

## 6. Membuat kolom baru

```python
# Membuat kolom dep_delay dalam satuan detik
flights['dep_delay_second'] = flights['dep_delay'] * 60
flights
```

```python
# Menggunakan if else
flights['semester'] = np.where(flights['month'] <= 6, 1,
                      np.where(flights['month'] >= 7, 2, 0))

flights
```

```python
# Menghapus kolom
flights.drop(columns=['dep_delay_second', 'semester'])
```
